```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
```

### ЗАВДАННЯ 1:

```python
data_d = pd.read_csv('data.csv')
data_d.head()
```

```python
inertia_d = []
for i in range(1, 11):
    claster = KMeans(n_clusters=i, n_init='auto').fit(data_d)
    inertia_d.append(claster.inertia_)
plt.grid()
plt.plot(range(1, 11), inertia_d)
plt.show()
```

```python
model_d = KMeans(n_clusters=2, n_init = 'auto').fit(data_d)
```

```python
data_mnist = pd.read_csv('mnist.csv')
data_mnist
```

```python
inertia_mnist = []
for i in range(1, 11):
    claster = KMeans(n_clusters=i, n_init='auto').fit(data_mnist)
    inertia_mnist.append(claster.inertia_)
plt.grid()
plt.plot(range(1, 11), inertia_mnist)
plt.show()
```

```python
model_mnist = KMeans(n_clusters=3, n_init = 'auto').fit(data_mnist)
```
